```python
 def top_pearson_correlation (df,n):
    corr_matrix = df.corr(method ='pearson')
    correlation = (corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
                 .stack()
                 .sort_values(ascending=False))
    correlation = pd.DataFrame(correlation).reset_index()
    correlation.columns=["Variable_1","Variable_2","Pearson_Correlation"]
    correlation = correlation.reindex(correlation.Pearson_Correlation.abs().sort_values(ascending=False).index).reset_index().drop(["index"],axis=1)
    return correlation.head(n)

PC=top_pearson_correlation(df2,20)
```

Adjust the number in the function call to get the required number of top correlation values

```python
PC
```

```python
def top_spearman_correlation (df,n):
    corr_matrix = df.corr(method ='spearman')
    correlation = (corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
                 .stack()
                 .sort_values(ascending=False))
    correlation = pd.DataFrame(correlation).reset_index()
    correlation.columns=["Variable_1","Variable_2","Spearman_Correlation"]
    correlation = correlation.reindex(correlation.Spearman_Correlation.sort_values(ascending=False).index).reset_index().drop(["index"],axis=1)
    return correlation.head(n)

SMC=top_spearman_correlation(df2,20)
```

```python
SMC
```

From the above two tables we observe that Pearson correlation coefficients measure only linear relationships. Spearman correlation coefficients measure only monotonic relationships. we can also say that Pearson correlation can't handle outliers whereas spearman can handle them.

```python
from scipy.stats import pearsonr
df2 = df2.fillna(df2.mean())
```

```python
from mlxtend.evaluate import permutation_test
import random
for i in range (100):
    x = df2[PC_100['Variable_1'][i]]
    y = df2[PC_100['Variable_2'][i]]
    observed_pearson = pearsonr(x,y)
    rank = 1
    for n in range(1000):
        z = np.random.permutation(y)
        score = pearsonr(x,z)
        if abs(score[0]) > abs(observed_pearson[0]):
                rank = rank +1
    pvalue = rank/10
    print ("Variable_1 = ", PC_100['Variable_1'][i], "Variable_2 = ", PC_100['Variable_2'][i],"Pvalue = ", pvalue)
        
```

Let's calculate similarity indices between all the countries based on natural gas - production,consumption,exports,imoprts,proved reserves

```python
from sklearn.neighbors import NearestNeighbors
gdp_cols=["GDP (purchasing power parity)","GDP - real growth rate","GDP - per capita (PPP)","Gross national saving","Exports","Imports"]
X = df2[gdp_cols]
X_norm=sklearn.preprocessing.normalize(X, norm='l2')
nbrs = NearestNeighbors(n_neighbors=197, algorithm='kd_tree').fit(X_norm)
distances, indices = nbrs.kneighbors(X_norm)
df_indices= pd.DataFrame(indices)
df_indices
```

```python
df_sim = pd.DataFrame(columns = ['Country', 'Most Similar', 'Least similar'])
for i in range(197):
    c = df2['Country Name'][i]
    c1 = df2['Country Name'][indices[i][1]]
    c2 = df2['Country Name'][indices[i][196]]
    df_sim = df_sim.append({'Country': c, 'Most Similar': c1, 'Least similar': c2}, ignore_index=True)
df_sim.head(20)
```
